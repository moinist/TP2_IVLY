```python
# 'first_join_date' 열을 Datetime 형식으로 변환
customer['first_join_date'] = pd.to_datetime(customer['first_join_date'])

# 'first_join_date' 열에서 연도 정보 추출하여 새로운 열 생성
customer['join_year'] = customer['first_join_date'].dt.year

# 연도별로 분류된 결과 확인
yearly_counts = customer['join_year'].value_counts().sort_index()

# 가입연도별 분포 막대 그래프 시각화
plt.figure(figsize=(8, 6))
yearly_counts.plot(kind='line')
plt.title("Join Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
# 전체 기간 내에서의 월별 가입 분포 계산
monthly_counts = customer['first_join_date'].dt.to_period('M').value_counts().sort_index()

# 월별 분포 막대 그래프 시각화
plt.figure(figsize=(25, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title("Monthly Join Distribution Over the Period")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 'birtdate' 열을 Datetime 형식으로 변환
customer['birthdate'] = pd.to_datetime(customer['birthdate'])

# 'first_join_date' 열에서 연도 정보 추출하여 새로운 열 생성
customer['birth_year'] = customer['birthdate'].dt.year

# 연도별로 분류된 결과 확인
birth_counts = customer['birth_year'].value_counts().sort_index()

# 가입연도별 분포 막대 그래프 시각화
plt.figure(figsize=(40, 20))
birth_counts.plot(kind='bar')
plt.title("Birth Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
transaction['promo_rate'] = transaction['promo_amount']/transaction['item_price']
promo_rate_means = (transaction.groupby(transaction['promo_code']).promo_rate.mean())*100

# 시각화
plt.figure(figsize=(12, 6))
promo_rate_means.plot(kind='bar')
plt.xlabel("Promotion Code")
plt.ylabel("Average Promotion Rate(%)")
plt.title("Average Promotion Rate by Promotion Code")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
ramadan = transaction[transaction['created_at'].between('2022-04-01', '2022-05-01', inclusive='both')]
promotion_counts = ramadan['promo_code'].value_counts()
promo_rate_means = (ramadan.groupby(ramadan['promo_code']).promo_rate.mean())*100
print(ramadan.dropna().promo_rate.mean())

# 시각화
plt.figure(figsize=(12, 6))
promo_rate_means.plot(kind='bar')
plt.xlabel("Promotion Code")
plt.ylabel("Average Promotion Rate(%)")
plt.title("Average Promotion Rate by Promotion Code")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(promotion_counts, labels=promotion_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("2022 Ramadan Promotion Code Distribution")
plt.show()

# promo_amount 컬럼의 값이 0이 아닌 행의 개수를 세기
non_zero_promo_count = ramadan[ramadan["promo_amount"] != 0].shape[0]

# 전체 행의 개수를 구하기
total_rows = ramadan.shape[0]

# 0이 아닌 promo_amount 행의 비율 계산
non_zero_promo_ratio = non_zero_promo_count / total_rows

# 시각화
labels = ['Non-Zero Promo Amount', 'Zero Promo Amount']
sizes = [non_zero_promo_ratio, 1 - non_zero_promo_ratio]
colors = ['#66b3ff', '#99ff99']
explode = (0.1, 0)  # Non-Zero Promo Amount 부분을 조금 돌출해서 강조

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 원형 그래프 유지
plt.title('Non-Zero Promo Amount Ratio')
plt.show()
```

```python
from datetime import datetime

# created_at 컬럼을 datetime 형식으로 변환
transaction["created_at"] = pd.to_datetime(transaction["created_at"])

# 월별로 그룹화하여 카운트 계산
monthly_counts = transaction.groupby(transaction["created_at"].dt.to_period("M")).size()

# 시각화
plt.figure(figsize=(20, 6))
monthly_counts.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Transaction Count")
plt.title("Monthly Transaction Distribution")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 제품별로 판매량과 수익을 계산
transaction["revenue"] = transaction["quantity"] * transaction["item_price"]
product_stats = transaction.groupby("product_id").agg({"quantity": "sum", "revenue": "sum"})

# 판매량을 기준으로 가장 많이 팔린 제품 상위 10위 데이터프레임 생성
top_sold_products = product_stats.nlargest(10, "quantity")

# 수익을 기준으로 가장 큰 수익을 낸 제품 상위 10위 데이터프레임 생성
top_revenue_products = product_stats.nlargest(10, "revenue")

print("Most Sold Products (Top 10):\n", top_sold_products)
print("\nHighest Revenue Products (Top 10):\n", top_revenue_products)
```

```python
# 2022년 이후 이탈 고객
customer_ids_before_2022 = transaction[transaction["created_at"].dt.year < 2022]["customer_id"]
customer_ids_after_2022 = transaction[transaction["created_at"].dt.year >= 2022]["customer_id"]
missing_customer_ids = customer_ids_before_2022[~customer_ids_before_2022.isin(customer_ids_after_2022)]

print("Customer IDs with transactions:\n", len(transaction.customer_id.unique()))
print("Customer IDs with transactions before 2022:\n", len(customer_ids_before_2022.unique()))
print("Customer IDs with transactions before 2022 but not after 2022:\n", len(missing_customer_ids.unique()))
```

```python
# 2022년 이후 이탈 고객 거래내역 데이터셋
missing_transaction = transaction[transaction["customer_id"].isin(missing_customer_ids.unique())]
missing_transaction
```

```python
# promo_amount 컬럼의 값이 0이 아닌 행의 개수를 세기
non_zero_promo_count = missing_transaction[missing_transaction["promo_amount"] != 0].shape[0]

# 전체 행의 개수를 구하기
total_rows = missing_transaction.shape[0]

# 0이 아닌 promo_amount 행의 비율 계산
non_zero_promo_ratio = non_zero_promo_count / total_rows

# 시각화
labels = ['Non-Zero Promo Amount', 'Zero Promo Amount']
sizes = [non_zero_promo_ratio, 1 - non_zero_promo_ratio]
colors = ['#66b3ff', '#99ff99']
explode = (0.1, 0)  # Non-Zero Promo Amount 부분을 조금 돌출해서 강조

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 원형 그래프 유지
plt.title('Non-Zero Promo Amount Ratio')
plt.show()
```
